## Task 1: Nested If-Else

```python
class_day = input("Aj class ka din hai? (Y/N): ").lower()

if class_day == "y":
    card = input("Card sath lai ho ya nahi? (Y/N): ").lower()

    if card == "y":
        fees = input("Fees pay karwai? (Y/N): ").lower()

        if fees == "y":
            assignment = input("Assignment complete karkay lai ho? (Y/N): ").lower()

            if assignment == "y":
                print("Class lai sakte ho")
            else:
                print("Assignment complete nahi hai,  tum class nahi lai sakte")
        else:
            print("Fees pay q nahi karwai, tum class nahi lai sakte")
    else:
        print("Card nahi hai,  tum class nahi lai sakte")
else:
    print("Aaj class ka din nahi hai")
```

## Task 2: Marks and Grade System

```python
m1 = int(input("Subject 1 marks: "))
m2 = int(input("Subject 2 marks: "))
m3 = int(input("Subject 3 marks: "))
m4 = int(input("Subject 4 marks: "))
m5 = int(input("Subject 5 marks: "))

total = m1 + m2 + m3 + m4 + m5
per = total / 500 * 100

if m1>=50 and m2>=50 and m3>=50 and m4>=50 and m5>=50:
    print("Pass")

    if per >= 80:
        print("Grade A+")
    elif per >= 70:
        print("Grade A")
    elif per >= 60:
        print("Grade B")
    else:
        print("Grade C")
else:
    print("Fail")
```
